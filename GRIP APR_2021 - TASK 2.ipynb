# Data Modelling

```python
wcss = []

for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The elbow curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

**The representation shows the elbow at the point 3, means we can choose k=3 clusters for the analysis.**



**Training the model and predicting for the same**

```python
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
ykmeans = km.fit_predict(x)
ykmeans
```

```python
km.cluster_centers_[:, 0]
```

```python
km.cluster_centers_[:,1]
```

# Visualization of clusters along centroids in the given data

```python
plt.figure(figsize=(6,6))
plt.scatter(x[ykmeans == 0, 0], x[ykmeans == 0, 1], s=100, c='red', marker='^', label='Iris-setosa')
plt.scatter(x[ykmeans == 1, 0], x[ykmeans == 1, 1], s=100, c='blue', marker='s', label='Iris-versicolour')
plt.scatter(x[ykmeans == 2, 0], x[ykmeans == 2, 1], s=100, c='yellow', marker="o", label='Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:,1], s = 100, c = 'maroon', label = 'Centroids',
            marker='.')
plt.title("Iris Data clusters along with centroids", fontsize=17)
plt.legend()
```

**Hence we can see the optimum number of clusters and visualizations of "iris" dataset**
